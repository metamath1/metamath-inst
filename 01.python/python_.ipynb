## 연산자 연습

````python
# ============================================================
# 연산자를 이용한 웨이퍼 생산제어 프로그램 작성
# ============================================================
# 이 예제는 실제 반도체 장비를 제어하는 프로그램이 아닙니다.
# 생산 진행 여부를 판단하기 전에 필요한 계산값을
# 연산자로 구해보는 교육용 예제입니다.
#
# 아직 조건문을 배우기 전 단계이므로 if문은 사용하지 않습니다.
# 대신 비교 연산자와 논리 연산자의 결과인 True/False를 확인합니다.
#
# 학습 목표:
# - 산술 연산자로 전체 칩 수, 정상 칩 수, 수율을 계산한다.
# - 비교 연산자로 공정값이 기준을 만족하는지 확인한다.
# - 논리 연산자 and로 여러 조건을 하나로 묶는다.
#
# 주의:
# 아래 기준값은 실제 공정 기준이 아닙니다.
# 파이썬 연산자 학습을 위한 교육용 가상 기준입니다.
# ============================================================

# ------------------------------------------------------------
# 교육용 가상 생산 데이터
# 실제 장비에서 가져온 데이터가 아닙니다.
# ------------------------------------------------------------
wafer_count = 3
chips_per_wafer = 650
defective_chips = 42

temperature = 73.5
pressure = 1.8

# ------------------------------------------------------------
# 교육용 가상 기준값
# 실제 공정 기준이 아닙니다.
# ------------------------------------------------------------
min_temperature = 70.0
max_temperature = 75.0

min_pressure = 1.5
max_pressure = 2.0

min_yield_rate = 95.0


# ============================================================
# 수강생 작성 영역
#
# 목표:
# 연산자를 사용해 아래 값을 계산하세요.
#
# 1. total_chips
#    전체 칩 수 = 웨이퍼 개수 * 웨이퍼 1장당 칩 개수
#
# 2. good_chips
#    정상 칩 수 = 전체 칩 수 - 불량 칩 수
#
# 3. defect_rate
#    불량률 = 불량 칩 수 / 전체 칩 수 * 100
#
# 4. yield_rate
#    수율 = 정상 칩 수 / 전체 칩 수 * 100
#
# 5. temperature_ok
#    온도가 min_temperature 이상 max_temperature 이하이면 True
#
# 6. pressure_ok
#    압력이 min_pressure 이상 max_pressure 이하이면 True
#
# 7. yield_ok
#    수율이 min_yield_rate 이상이면 True
#
# 8. production_ok
#    온도, 압력, 수율 조건이 모두 True이면 True
#
# 주의:
# 이 예제에서는 if문을 사용하지 않습니다.
# ============================================================

###########################################################################
# 여기에 코드 작성
total_chips = wafer_count * chips_per_wafer
good_chips = total_chips - defective_chips

defect_rate = defective_chips / total_chips * 100
yield_rate = good_chips / total_chips * 100

temperature_ok = temperature >= min_temperature and temperature <= max_temperature
pressure_ok = pressure >= min_pressure and pressure <= max_pressure
yield_ok = yield_rate >= min_yield_rate

production_ok = temperature_ok and pressure_ok and yield_ok
###########################################################################



print("[웨이퍼 생산 상태 계산 결과]")
print("웨이퍼 개수:", wafer_count)
print("웨이퍼 1장당 칩 개수:", chips_per_wafer)

print("전체 칩 수:", total_chips)
print("정상 칩 수:", good_chips)
print("불량 칩 수:", defective_chips)
print("불량률:", round(defect_rate, 2), "%")
print("수율:", round(yield_rate, 2), "%")
print("온도 기준 만족:", temperature_ok)
print("압력 기준 만족:", pressure_ok)
print("수율 기준 만족:", yield_ok)
print("생산 진행 조건 만족:", production_ok)
````

## 자료구조 연습

````python
# ============================================================
# 웨이퍼 재고량 CRUD하는 프로그램
# ============================================================
# 이 예제는 실제 반도체 재고 시스템을 그대로 구현한 것이 아닙니다.
# 파이썬의 dict 자료형을 연습하기 위해
# "웨이퍼 ID별 재고 관리" 상황을 단순화한 교육용 예제입니다.
#
# 학습 목표:
# - dict 자료형의 key-value 구조를 이해한다.
# - 데이터를 추가(Create), 조회(Read), 수정(Update), 삭제(Delete)한다.
# - 웨이퍼 ID처럼 고유한 식별자를 key로 사용하는 방식을 익힌다.
#
# 수강생은 "수강생 작성 영역"만 작성합니다.
# 메뉴 출력, 입력 처리, 반복 실행 코드는 과제 진행을 위한 제공 코드입니다.
# ============================================================


inventory = {
    "WAFER-001": {"quantity": 25, "status": "입고완료"},
    "WAFER-002": {"quantity": 40, "status": "검사대기"},
    "WAFER-003": {"quantity": 15, "status": "생산투입가능"}
}


def show_all_inventory():
    # 전체 재고를 보기 좋게 출력하기 위한 제공 함수입니다.
    # 함수 문법을 아직 자세히 배우지 않았더라도 이 부분은 수정하지 않습니다.
    print("\n[전체 웨이퍼 재고]")

    if len(inventory) == 0:
        print("등록된 웨이퍼가 없습니다.")
        return

    for wafer_id, info in inventory.items():
        print(
            wafer_id,
            "| 수량:", info["quantity"],
            "| 상태:", info["status"]
        )


while True:
    print("\n==============================")
    print("웨이퍼 재고 관리 프로그램")
    print("==============================")
    print("1. 전체 재고 조회")
    print("2. 신규 웨이퍼 등록")
    print("3. 특정 웨이퍼 조회")
    print("4. 웨이퍼 정보 수정")
    print("5. 웨이퍼 삭제")
    print("0. 종료")

    menu = input("메뉴를 선택하세요: ")

    if menu == "1":
        show_all_inventory()

    elif menu == "2":
        wafer_id = input("등록할 웨이퍼 ID를 입력하세요: ")
        quantity = int(input("수량을 입력하세요: "))
        status = input("상태를 입력하세요: ")

        # ====================================================
        # 수강생 작성 영역: Create
        #
        # 목표:
        # 입력받은 wafer_id, quantity, status를 사용하여
        # inventory에 새로운 웨이퍼 정보를 등록하세요.
        # ====================================================

        ###########################################################################
        # 여기에 코드 작성
        # Create: 신규 웨이퍼 등록
        inventory[wafer_id] = {
            "quantity": quantity,
            "status": status
        }
        ###########################################################################

        print("등록 완료")

    elif menu == "3":
        wafer_id = input("조회할 웨이퍼 ID를 입력하세요: ")

        # ====================================================
        # 수강생 작성 영역: Read
        #
        # 목표:
        # 입력받은 wafer_id가 inventory에 존재하면
        # 해당 웨이퍼 정보를 출력하세요.
        #
        # 존재하지 않으면
        # "해당 웨이퍼 ID가 없습니다."를 출력하세요.
        # ====================================================

        ###########################################################################
        # 여기에 코드 작성
        # Read: 특정 웨이퍼 조회
        if wafer_id in inventory:
            print(wafer_id, inventory[wafer_id])
        else:
            print("해당 웨이퍼 ID가 없습니다.")
        ###########################################################################

    elif menu == "4":
        wafer_id = input("수정할 웨이퍼 ID를 입력하세요: ")
        quantity = int(input("수정할 수량을 입력하세요: "))
        status = input("수정할 상태를 입력하세요: ")

        # ====================================================
        # 수강생 작성 영역: Update
        #
        # 목표:
        # wafer_id가 존재하면 해당 웨이퍼의 quantity와 status를
        # 입력받은 값으로 수정하세요.
        # ====================================================

        ###########################################################################
        # 여기에 코드 작성
        # Update: 웨이퍼 정보 수정
        if wafer_id in inventory:
            inventory[wafer_id]["quantity"] = quantity
            inventory[wafer_id]["status"] = status
            print("수정 완료")
        else:
            print("해당 웨이퍼 ID가 없습니다.")
        ###########################################################################

    elif menu == "5":
        wafer_id = input("삭제할 웨이퍼 ID를 입력하세요: ")

        # ====================================================
        # 수강생 작성 영역: Delete
        #
        # 목표:
        # wafer_id가 존재하면 inventory에서 해당 데이터를 삭제하세요.
        # ====================================================

        ###########################################################################
        # 여기에 코드 작성
        # Delete: 웨이퍼 정보 삭제
        if wafer_id in inventory:
            del inventory[wafer_id]
            print("삭제 완료")
        else:
            print("해당 웨이퍼 ID가 없습니다.")
        ###########################################################################

    elif menu == "0":
        print("프로그램을 종료합니다.")
        break

    else:
        print("잘못된 메뉴입니다. 다시 선택하세요.")
````

- BMI 계산기

- <font color="red">**프롬프트**</font>:
```
비만도를 출력하는 if문을 사용하는 간단한 프로그램
BMI = weight / height^2
저체중 BMI < 18.5
정상: 18.5 <= BMI < 23
과체중: 23 <= BMI < 25
비만 25 <= BMI
```

````python
weight = float(input("몸무게(kg): "))
height = float(input("키(m): "))

# BMI 계산
bmi = weight / (height ** 2)

print("BMI =", round(bmi, 1))

# 비만도 판정
if bmi < 18.5:
    print("저체중")
elif bmi < 23:
    print("정상")
elif bmi < 25:
    print("과체중")
else:
    print("비만")
````

- 행렬 곱셈

- <font color="red">**프롬프트**</font>: 3x3 두 행렬을 for 루프를 이용해서  곱셈하는 코드

````python
# 3x3 행렬
A = [[1,0,1], [0,2,0], [1,2,1]]
B = [[2,3,1], [0,1,1], [1,1,1]]

# 결과 행렬 초기화
result = [
    [0, 0, 0],
    [0, 0, 0],
    [0, 0, 0]
]

# 행렬 곱셈
for i in range(3):          # A의 행
    for j in range(3):      # B의 열
        for k in range(3):  # 곱셈 계산
            result[i][j] += A[i][k] * B[k][j]

# 결과 출력
print("결과 행렬")

for row in result:
    print(row)
````

## 함수 활용 연습

````python
# ============================================================
# 웨이퍼 개수와 불량칩의 개수를 입력하면
# 수율을 계산하는 프로그램
# ============================================================
# 이 예제는 실제 생산 시스템에서 데이터를 가져오는 것이 아닙니다.
# 웨이퍼 개수, 웨이퍼당 칩 개수, 불량 칩 개수가
# 1초마다 들어오는 상황을 단순화한 교육용 예제입니다.
#
# 수율 계산식 역시 문법 학습을 위해 단순화한 계산식입니다.
# 실제 현업에서는 공정, 제품, 검사 기준, 재작업 여부 등에 따라
# 수율 정의가 달라질 수 있습니다.
#
# 학습 목표:
# - 함수를 정의한다.
# - 매개변수를 사용한다.
# - 계산 결과를 return 한다.
# - 반복적으로 들어오는 데이터에 같은 함수를 재사용한다.
#
# 수강생은 calculate_yield() 함수 내부만 작성합니다.
# random과 time은 실시간 데이터처럼 보이게 하기 위한 제공 코드입니다.
# ============================================================

import random
import time


def calculate_yield(wafer_count, chips_per_wafer, defective_chips):
    # ========================================================
    # 수강생 작성 영역
    #
    # 목표:
    # 웨이퍼 개수, 웨이퍼 1장당 칩 개수, 불량 칩 개수를 사용해
    # 수율을 계산하고 return 하세요.
    #
    # 교육용 단순 계산식:
    # 전체 칩 수 = 웨이퍼 개수 * 웨이퍼 1장당 칩 개수
    # 정상 칩 수 = 전체 칩 수 - 불량 칩 수
    # 수율 = 정상 칩 수 / 전체 칩 수 * 100
    #
    # 주의:
    # 이 계산식은 파이썬 함수 학습을 위한 단순화된 수율 계산입니다.
    # 실제 제조 현장의 수율 계산은 제품/공정/검사 기준에 따라
    # 더 복잡할 수 있습니다.
    # ========================================================

    ###########################################################################
    # 여기에 코드 작성
    total_chips = wafer_count * chips_per_wafer
    good_chips = total_chips - defective_chips
    yield_rate = good_chips / total_chips * 100

    return yield_rate
    ###########################################################################

    # 아래 return 0은 코드가 실행되도록 넣어둔 임시값입니다.
    # 과제 수행 시 실제 계산 결과를 return 하도록 수정하세요.
    return 0


print("[실시간 웨이퍼 수율 계산 프로그램]")
print("1초마다 생산 데이터가 들어온다고 가정합니다.")
print("아래 데이터는 파이썬 문법 학습을 위한 교육용 가상 데이터입니다.\n")


for i in range(10):
    # --------------------------------------------------------
    # 교육용 가상 생산 데이터 생성
    # 실제 생산 설비에서 가져온 데이터가 아닙니다.
    # --------------------------------------------------------
    wafer_count = random.randint(1, 5)
    chips_per_wafer = random.randint(500, 800)

    total_chips = wafer_count * chips_per_wafer

    # 전체 칩 수의 최대 약 8% 범위 안에서 불량 칩 개수를 임의 생성합니다.
    # 이 값 역시 실제 불량률을 의미하지 않는 교육용 가상값입니다.
    defective_chips = random.randint(0, int(total_chips * 0.08))

    yield_rate = calculate_yield(
        wafer_count,
        chips_per_wafer,
        defective_chips
    )

    print("측정 번호:", i + 1)
    print("웨이퍼 개수:", wafer_count)
    print("웨이퍼 1장당 칩 개수:", chips_per_wafer)
    print("불량 칩 개수:", defective_chips)
    print("수율:", round(yield_rate, 2), "%")
    print("-" * 50)

    time.sleep(1)
````

## Walk-through

- 로또 번호 생성기

- <font color="red">**프롬프트**</font>:

```
호출하면 로또 번호 6개를 반환하는 함수 작성
사용자로 부터 게임수를 입력받고 입력 받은 수 만큼 함수를 반복 호출하여 여러개의 번호를 생성하는 코드
```


````python
import random


# 로또 번호 생성 함수
def make_lotto():
    numbers = random.sample(range(1, 46), 6)
    numbers.sort()
    return numbers


# 게임 수 입력
count = int(input("게임 수 입력: "))

# 반복 생성
for i in range(count):
    lotto = make_lotto()
    print(i + 1, "게임:", lotto)
````

- <font color="red">**프롬프트**</font>: sample 함수를 모른다고 가정하고 다시 작성

````python
import random


# 로또 번호 생성 함수
def make_lotto():
    lotto = []

    while len(lotto) < 6:
        number = random.randint(1, 45)
        # 중복 검사
        if number not in lotto:
            lotto.append(number)

    lotto.sort()
    return lotto


# 게임 수 입력
count = int(input("게임 수 입력: "))

# 여러 게임 생성
for i in range(count):
    numbers = make_lotto()
    print(i + 1, "게임:", numbers)
````

- 평균과 표준 편차

````python
import random

# 20행 5열의 2차원 리스트 생성 (1~100 사이의 랜덤 정수)
data = [[random.randint(20, 99) for _ in range(5)] for _ in range(20)]

# 결과 출력
for row in data:
    print(row)
````

````python
import math

rows = len(data)
cols = len(data[0])

means = []
std_devs = []

# 컬럼별로 통계 계산
for j in range(cols):
    # 1. 평균 계산
    col_sum = 0
    for i in range(rows):
        col_sum += data[i][j]
    mean_val = col_sum / rows
    means.append(mean_val)

    # 2. 표준 편차 계산
    variance_sum = 0
    for i in range(rows):
        variance_sum += (data[i][j] - mean_val) ** 2
    std_val = math.sqrt(variance_sum / rows)
    std_devs.append(std_val)

    print(f"Column {j+1}: Mean = {mean_val:.2f}, Std Dev = {std_val:.2f}")

print("\nMeans list:", means)
print("Std Devs list:", std_devs)
````
