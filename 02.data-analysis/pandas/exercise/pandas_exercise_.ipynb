- 수치형 컬럼과 문자형 컬럼의 기초 통계량을 확인하세요.
- `Particle_Count / Cycle_Time`을 계산하여 `Particle_per_Cycle` 컬럼을 추가하세요.
- `Particle_per_Cycle` 값이 큰 순서로 상위 3개 행을 확인하세요. (인덱스 번호 2949, 2919, 1747이 나와야함)

## 문제 5. 결측치 처리

`df`에 대해서 결측치가 있는 컬럼을 확인한 뒤, 다음 방식으로 결측치를 처리하세요. 수치형 컬럼은 `num_cols`, 문자형 컬럼은 `cat_cols`를 사용하세요.

- 전체 결측치의 개수를 확인하세요.
- 수치형 컬럼의 결측치는 각 컬럼의 중앙값으로 채우세요.
- 문자형 컬럼의 결측치는 `"Unknown"`으로 채우세요.
- 결측치 처리 후 결측치가 남아 있는지 다시 확인하세요.

```python
df_clean = df.copy()
num_cols = df_clean.select_dtypes(include="number").columns
cat_cols = df_clean.select_dtypes(exclude="number").columns

print("결측치 총합:", df.isnull().sum().sum())

df_clean[num_cols] = df_clean[num_cols].fillna(df_clean[num_cols].median())
df_clean[cat_cols] = df_clean[cat_cols].fillna("Unknown")

print("남은 결측치 총합:", df_clean.isna().sum().sum())
```

## 문제 6. IQR 방식으로 이상치 찾기와 제거하기

`df_clean`에 대해서  `Particle_Count` 컬럼을 대상으로 IQR 방식의 이상치 기준을 계산하세요.

다음 값을 구하세요.

- Q1
- Q3
- IQR
- lower bound
- upper bound

그다음 다음 작업을 수행하세요.

- `Particle_Count`가 IQR 기준을 벗어난 행을 찾으세요.
- 이상치 개수를 출력하세요.
- 이상치를 제거한 결과를 `df_no_outlier`에 저장하세요. `df_no_outlier`의 행수는 3539가 되야함

```python
col = "Particle_Count"

Q1 = df_clean[col].quantile(0.25)
Q3 = df_clean[col].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print("Q1:", Q1)
print("Q3:", Q3)
print("IQR:", IQR)
print("lower bound:", lower_bound)
print("upper bound:", upper_bound)

outlier_cond = (df_clean[col] < lower_bound) | (df_clean[col] > upper_bound)
outlier_df = df_clean[outlier_cond]

print("이상치 개수:", len(outlier_df))
display(outlier_df.head())

df_no_outlier = df_clean[~outlier_cond].copy()
print("이상치 제거 후 데이터 크기:", df_no_outlier.shape)
```

## 문제 7. 공장별 불량률 계산

`df_clean`에 대해서 공장별(`Factory`)로 다음 값을 계산하세요.

- 생산 건수
- 불량 건수
- 불량률

계산한 정보를 담은 `df_factory` 데이터프레임에 대해서 불량률이 높은 순서로 정렬하세요.

```python
df_size = df_clean.groupby('Factory').size()
df_defect = df_clean.groupby('Factory')['Defect'].sum()
df_defect_rate = df_clean.groupby('Factory')['Defect'].mean()

display(df_size)
display(df_defect)
display(df_defect_rate)

df_factory = pd.concat([df_size,df_defect,df_defect_rate], axis=1)
df_factory.columns = ['Count', 'Defect Count', 'Defect Rate']
df_factory.sort_values('Defect Rate', ascending=False)
```

```python
factory_summary = (
    df_clean.groupby("Factory")
    .agg(
        Production_Count=("Defect", "size"),
        Defect_Count=("Defect", "sum"),
        Defect_Rate=("Defect", "mean")
    )
    .sort_values("Defect_Rate", ascending=False)
)

factory_summary["Defect_Rate_Percent"] = factory_summary["Defect_Rate"] * 100
display(factory_summary)
```

## 문제 8. 공장 × 교대조, 공장 × 제품유형 불량률 계산

다음 작업을 수행하세요.

- `pivot_table()`을 사용하여 공장 × 교대조별 불량률을 계산하여 `factory_shift_rate`에 저장하고 확인하세요.
- `pivot_table()`을 사용하여 공장 × 제품유형별 불량률을 계산여 `factory_product_rate`에 저장하고 확인하세요.
- `crosstab()`을 사용하여 공장별 제품유형 생산 건수를 `factory_product_count`에 저장하고 확인하세요.

```python
factory_shift_rate = pd.pivot_table(
    df_clean,
    index="Factory",
    columns="Shift",
    values="Defect",
    aggfunc="mean"
)
display(factory_shift_rate)

factory_product_rate = pd.pivot_table(
    df_clean,
    index="Factory",
    columns="Product_Type",
    values="Defect",
    aggfunc="mean"
)
display(factory_product_rate)

factory_product_count = pd.crosstab(df_clean["Factory"], df_clean["Product_Type"])
display(factory_product_count)
```
