# [System Instruction: Data Analysis Master Mentor]

## 1. Persona & Expert Mindset

- **기본 환경:** 모든 데이터 탐색, 시각화, 전처리, 모델링 과정은 사용자의 컴퓨터에서 수행되어야 한다.

## 1단계: 데이터 구조 파악

````python
# 1단계: 데이터 구조 파악
# 목적:
# - 전체 데이터의 크기, 컬럼 구조, 결측치, 타깃 분포를 확인한다.
# - 식별자 컬럼과 실제 공정 변수 컬럼을 분리한다.
# - Machine_Status가 분석 가능한 수준으로 포함되어 있는지 확인한다.

import pandas as pd
import numpy as np

# 실제 파일명에 맞게 수정하세요.
file_path = "DieCasting_Raw_Data.csv"

df = pd.read_csv(file_path)

print("데이터 크기:", df.shape)

print("\n컬럼 목록:")
print(df.columns.tolist())

print("\n처음 5행:")
display(df.head())

print("\n마지막 5행:")
display(df.tail())

print("\n데이터 타입:")
display(df.dtypes.to_frame("dtype"))

# 타깃 컬럼
target_col = "Machine_Status"

# 식별자 후보 컬럼
id_cols = [col for col in ["Shot", "_id"] if col in df.columns]

# 실제 분석용 피처 후보
feature_cols = [
    col for col in df.columns
    if col not in id_cols + [target_col]
]

print("\n식별자 컬럼:", id_cols)
print("분석용 피처 컬럼:", feature_cols)

# 결측치 요약
missing_summary = pd.DataFrame({
    "missing_count": df.isna().sum(),
    "missing_rate_percent": (df.isna().mean() * 100).round(2)
}).sort_values("missing_count", ascending=False)

print("\n결측치 요약:")
display(missing_summary)

# 타깃 분포
print("\nMachine_Status 분포:")
target_dist = pd.DataFrame({
    "count": df[target_col].value_counts(dropna=False),
    "rate_percent": (df[target_col].value_counts(dropna=False, normalize=True) * 100).round(2)
})
display(target_dist)

# 컬럼별 고유값 개수
nunique_summary = pd.DataFrame({
    "n_unique": df.nunique(dropna=False)
}).sort_values("n_unique")

print("\n컬럼별 고유값 개수:")
display(nunique_summary)

# 수치형 변수 기초 통계
print("\n수치형 변수 기초 통계:")
display(
    df[feature_cols]
    .describe(percentiles=[0.01, 0.05, 0.25, 0.5, 0.75, 0.95, 0.99])
    .T
)

# 중복 패턴 확인
print("\n전체 행 기준 완전 중복 행 수:", df.duplicated().sum())

if len(feature_cols) > 0:
    print(
        "식별자 제외 기준 중복 행 수:",
        df.duplicated(subset=feature_cols + [target_col]).sum()
    )

# 타깃이 0과 1을 모두 포함할 때만 상관 확인
if df[target_col].nunique(dropna=False) > 1:
    corr_with_target = (
        df[feature_cols + [target_col]]
        .corr(numeric_only=True)[target_col]
        .drop(target_col)
        .sort_values(key=lambda x: x.abs(), ascending=False)
    )

    print("\nMachine_Status와의 상관계수:")
    display(corr_with_target.to_frame("corr_with_Machine_Status"))
else:
    print("\n현재 데이터에서는 Machine_Status가 한 종류뿐이므로 타깃 상관분석을 건너뜁니다.")
````

## 2단계: 결측 진단

````python
# 2단계: 결측 진단
# 목적:
# - 결측이 특정 행에 몰려 있는지 확인한다.
# - 공정 변수 14개가 동시에 결측인지 확인한다.
# - 결측 자체가 Machine_Status=1과 관련 있는지 확인한다.
# - Missing Map으로 결측 패턴을 시각적으로 확인한다.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

file_path = "DieCasting_Raw_Data.csv"
df = pd.read_csv(file_path)

target_col = "Machine_Status"
id_cols = ["Shot", "_id"]

feature_cols = [
    col for col in df.columns
    if col not in id_cols + [target_col]
]

# 행 단위 결측 개수
df["row_missing_count"] = df[feature_cols].isna().sum(axis=1)

# 공정 변수 전체 결측 여부
df["all_process_missing"] = df[feature_cols].isna().all(axis=1)

# 공정 변수 일부라도 결측 여부
df["any_process_missing"] = df[feature_cols].isna().any(axis=1)

print("공정 변수 개수:", len(feature_cols))

print("\n행 단위 공정 변수 결측 개수 분포:")
display(
    df["row_missing_count"]
    .value_counts()
    .sort_index()
    .to_frame("row_count")
)

print("\n공정 변수 전체 결측 여부:")
display(
    df["all_process_missing"]
    .value_counts()
    .rename(index={False: "전체 결측 아님", True: "공정 변수 전체 결측"})
    .to_frame("row_count")
)

print("\n공정 변수 일부라도 결측 여부:")
display(
    df["any_process_missing"]
    .value_counts()
    .rename(index={False: "결측 없음", True: "결측 있음"})
    .to_frame("row_count")
)

print("\nMachine_Status 결측 여부와 공정 변수 결측 여부 교차표:")
status_missing_table = pd.crosstab(
    df["any_process_missing"],
    df[target_col].isna(),
    rownames=["공정 변수 결측 있음"],
    colnames=["Machine_Status 결측"]
)
display(status_missing_table)

# 라벨이 있는 데이터만 대상으로 결측 여부별 비정상 비율 확인
labeled = df[df[target_col].notna()].copy()

missing_target_rate = (
    labeled
    .groupby("any_process_missing")[target_col]
    .agg(["count", "sum", "mean"])
    .rename(columns={
        "count": "labeled_count",
        "sum": "abnormal_count",
        "mean": "abnormal_rate"
    })
)

missing_target_rate["abnormal_rate_percent"] = (
    missing_target_rate["abnormal_rate"] * 100
).round(2)

print("\n공정 변수 결측 여부별 비정상 비율:")
display(missing_target_rate)

# 결측이 몰린 구간 확인
print("\n공정 변수 결측이 있는 행의 Shot 구간 요약:")
missing_rows = df[df["any_process_missing"]]

display(
    missing_rows[["Shot", "_id", "row_missing_count", target_col]]
    .describe(include="all")
)

print("\n공정 변수 결측이 있는 행 일부:")
display(
    missing_rows[["Shot", "_id", "row_missing_count", target_col]]
    .head(20)
)

# Missing Map 시각화
# 너무 많은 컬럼을 한눈에 보기 위해 공정 변수 + 타깃만 표시
missing_matrix = df[feature_cols + [target_col]].isna()

plt.figure(figsize=(12, 6))
plt.imshow(missing_matrix.T, aspect="auto", interpolation="nearest")
plt.yticks(range(len(missing_matrix.columns)), missing_matrix.columns)
plt.xlabel("Row index")
plt.ylabel("Column")
plt.title("Missing Map: Process Variables and Machine_Status")
plt.tight_layout()
plt.show()

# 행별 결측 개수 시각화
plt.figure(figsize=(12, 4))
plt.plot(df.index, df["row_missing_count"])
plt.xlabel("Row index")
plt.ylabel("Number of missing process variables")
plt.title("Row-wise Missing Count")
plt.tight_layout()
plt.show()
````

## 2단계 결측 처리 방침 적용 코드

````python
# 2단계 결측 처리 방침 적용 코드
# 목적:
# - Machine_Status가 없는 행은 학습용 데이터에서 제외한다.
# - 공정 변수 결측 여부를 별도 파생 변수로 남긴다.
# - 공정 변수 14개가 모두 결측인 행은 별도 플래그로 관리한다.
# - 결측 처리 전용 정제 파일을 생성한다.

import pandas as pd
import numpy as np

file_path = "DieCasting_Raw_Data.csv"
df = pd.read_csv(file_path)

target_col = "Machine_Status"
id_cols = ["Shot", "_id"]

feature_cols = [
    col for col in df.columns
    if col not in id_cols + [target_col]
]

# 결측 플래그 생성
df["process_missing_count"] = df[feature_cols].isna().sum(axis=1)
df["process_missing_flag"] = df["process_missing_count"] > 0
df["all_process_missing_flag"] = df["process_missing_count"] == len(feature_cols)

# Machine_Status가 없는 행은 학습용에서 제외
df_labeled = df[df[target_col].notna()].copy()

# Machine_Status를 정수형으로 변환
df_labeled[target_col] = df_labeled[target_col].astype(int)

print("원본 데이터 크기:", df.shape)
print("라벨 있는 데이터 크기:", df_labeled.shape)

print("\n라벨 있는 데이터의 Machine_Status 분포:")
display(
    pd.DataFrame({
        "count": df_labeled[target_col].value_counts(),
        "rate_percent": (df_labeled[target_col].value_counts(normalize=True) * 100).round(2)
    })
)

print("\n라벨 있는 데이터의 결측 플래그별 비정상 비율:")
missing_flag_summary = (
    df_labeled
    .groupby(["process_missing_flag", "all_process_missing_flag"])[target_col]
    .agg(["count", "sum", "mean"])
    .rename(columns={
        "count": "row_count",
        "sum": "abnormal_count",
        "mean": "abnormal_rate"
    })
)

missing_flag_summary["abnormal_rate_percent"] = (
    missing_flag_summary["abnormal_rate"] * 100
).round(2)

display(missing_flag_summary)

# 공정 변수 전체 결측 행과 아닌 행 분리
df_labeled_complete_or_partial = df_labeled[
    ~df_labeled["all_process_missing_flag"]
].copy()

df_labeled_all_missing = df_labeled[
    df_labeled["all_process_missing_flag"]
].copy()

print("\n학습 후보 데이터 크기: 공정 변수 전체 결측 제외")
print(df_labeled_complete_or_partial.shape)

print("\n공정 변수 전체 결측이지만 라벨은 있는 행")
print(df_labeled_all_missing.shape)

# 정제 파일 저장
df_labeled.to_csv("diecasting_labeled_with_missing_flags.csv", index=False, encoding="utf-8-sig")
df_labeled_complete_or_partial.to_csv("diecasting_model_base.csv", index=False, encoding="utf-8-sig")
df_labeled_all_missing.to_csv("diecasting_all_process_missing_rows.csv", index=False, encoding="utf-8-sig")

print("\n저장 완료:")
print("- diecasting_labeled_with_missing_flags.csv")
print("- diecasting_model_base.csv")
print("- diecasting_all_process_missing_rows.csv")
````

## 3단계 EDA: 정상 vs 비정상 차이 비교

````python
# 3단계: 정상 vs 비정상 EDA
# 목적:
# - Machine_Status 0/1 간 주요 공정 변수의 분포 차이를 확인한다.
# - 변수별 효과크기(Cohen's d)를 계산한다.
# - 변수 구간별 비정상 비율을 확인한다.
# - 이후 피처 엔지니어링과 현장 Rule 후보를 만들기 위한 근거를 확보한다.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

file_path = "diecasting_model_base.csv"
df = pd.read_csv(file_path)

target_col = "Machine_Status"

# 분석 대상 주요 변수
focus_cols = [
    "High_Velocity",
    "Cylinder_Pressure",
    "Casting_Pressure",
    "Cycle_Time",
    "Rapid_Rise_Time"
]

print("EDA 대상 데이터 크기:", df.shape)

print("\nMachine_Status 분포:")
display(
    pd.DataFrame({
        "count": df[target_col].value_counts(),
        "rate_percent": (df[target_col].value_counts(normalize=True) * 100).round(2)
    })
)

# ------------------------------------------------------------
# 1. 정상/비정상별 기초 통계
# ------------------------------------------------------------
summary_list = []

for col in focus_cols:
    temp = (
        df
        .groupby(target_col)[col]
        .agg(["count", "mean", "std", "min", "median", "max"])
        .reset_index()
    )
    temp.insert(0, "variable", col)
    summary_list.append(temp)

group_summary = pd.concat(summary_list, ignore_index=True)

print("\n정상/비정상별 주요 변수 기초 통계:")
display(group_summary)

# ------------------------------------------------------------
# 2. 효과크기 계산: Cohen's d
# ------------------------------------------------------------
def cohens_d(x0, x1):
    x0 = x0.dropna()
    x1 = x1.dropna()
    
    n0 = len(x0)
    n1 = len(x1)
    
    if n0 < 2 or n1 < 2:
        return np.nan
    
    pooled_std = np.sqrt(
        ((n0 - 1) * x0.var() + (n1 - 1) * x1.var()) / (n0 + n1 - 2)
    )
    
    if pooled_std == 0:
        return np.nan
    
    return (x1.mean() - x0.mean()) / pooled_std

effect_rows = []

for col in focus_cols:
    normal = df.loc[df[target_col] == 0, col]
    abnormal = df.loc[df[target_col] == 1, col]
    
    effect_rows.append({
        "variable": col,
        "normal_mean": normal.mean(),
        "abnormal_mean": abnormal.mean(),
        "mean_diff_abnormal_minus_normal": abnormal.mean() - normal.mean(),
        "cohens_d": cohens_d(normal, abnormal)
    })

effect_df = pd.DataFrame(effect_rows)
effect_df["abs_cohens_d"] = effect_df["cohens_d"].abs()
effect_df = effect_df.sort_values("abs_cohens_d", ascending=False)

print("\n효과크기 기준 변수 중요도:")
display(effect_df)

# ------------------------------------------------------------
# 3. Boxplot: 정상 vs 비정상 분포 비교
# ------------------------------------------------------------
for col in focus_cols:
    plot_df = df[[target_col, col]].dropna()
    
    normal_values = plot_df.loc[plot_df[target_col] == 0, col]
    abnormal_values = plot_df.loc[plot_df[target_col] == 1, col]
    
    plt.figure(figsize=(6, 4))
    plt.boxplot(
        [normal_values, abnormal_values],
        labels=["Normal(0)", "Abnormal(1)"],
        showfliers=True
    )
    plt.title(f"{col} by Machine_Status")
    plt.ylabel(col)
    plt.tight_layout()
    plt.show()

# ------------------------------------------------------------
# 4. KDE: 정상 vs 비정상 분포 형태 비교
# ------------------------------------------------------------
# Cycle_Time은 극단값 6170 같은 값이 있으면 그래프가 눌릴 수 있으므로
# 시각화용으로만 99% 분위수에서 클리핑한 컬럼을 사용한다.
df_plot = df.copy()

cycle_clip_upper = df_plot["Cycle_Time"].quantile(0.99)
df_plot["Cycle_Time_for_plot"] = df_plot["Cycle_Time"].clip(upper=cycle_clip_upper)

plot_cols = [
    "High_Velocity",
    "Cylinder_Pressure",
    "Casting_Pressure",
    "Cycle_Time_for_plot",
    "Rapid_Rise_Time"
]

for col in plot_cols:
    plt.figure(figsize=(7, 4))
    
    df_plot.loc[df_plot[target_col] == 0, col].dropna().plot(
        kind="kde",
        label="Normal(0)"
    )
    df_plot.loc[df_plot[target_col] == 1, col].dropna().plot(
        kind="kde",
        label="Abnormal(1)"
    )
    
    plt.title(f"KDE Distribution: {col}")
    plt.xlabel(col)
    plt.legend()
    plt.tight_layout()
    plt.show()

# ------------------------------------------------------------
# 5. 분위 구간별 비정상 비율
# ------------------------------------------------------------
rate_tables = {}

for col in focus_cols:
    temp = df[[col, target_col]].dropna().copy()
    
    # 중복 경계가 있을 수 있으므로 duplicates='drop' 사용
    temp["bin"] = pd.qcut(temp[col], q=4, duplicates="drop")
    
    rate_table = (
        temp
        .groupby("bin", observed=True)[target_col]
        .agg(["count", "sum", "mean"])
        .rename(columns={
            "count": "row_count",
            "sum": "abnormal_count",
            "mean": "abnormal_rate"
        })
    )
    
    rate_table["abnormal_rate_percent"] = (
        rate_table["abnormal_rate"] * 100
    ).round(2)
    
    rate_tables[col] = rate_table
    
    print(f"\n[{col}] 분위 구간별 비정상 비율:")
    display(rate_table)

# ------------------------------------------------------------
# 6. 간단한 Rule 후보 테스트
# ------------------------------------------------------------
# 이전 상관분석 결과를 바탕으로,
# High_Velocity 저하, Cylinder_Pressure 저하, Casting_Pressure 저하,
# Cycle_Time 증가 조건을 단순 조합해본다.

hv_q25 = df["High_Velocity"].quantile(0.25)
cp_q25 = df["Cylinder_Pressure"].quantile(0.25)
cast_q25 = df["Casting_Pressure"].quantile(0.25)
ct_q75 = df["Cycle_Time"].quantile(0.75)

df["rule_low_high_velocity"] = df["High_Velocity"] <= hv_q25
df["rule_low_cylinder_pressure"] = df["Cylinder_Pressure"] <= cp_q25
df["rule_low_casting_pressure"] = df["Casting_Pressure"] <= cast_q25
df["rule_long_cycle_time"] = df["Cycle_Time"] >= ct_q75

df["rule_score"] = (
    df["rule_low_high_velocity"].astype(int)
    + df["rule_low_cylinder_pressure"].astype(int)
    + df["rule_low_casting_pressure"].astype(int)
    + df["rule_long_cycle_time"].astype(int)
)

rule_summary = (
    df
    .groupby("rule_score")[target_col]
    .agg(["count", "sum", "mean"])
    .rename(columns={
        "count": "row_count",
        "sum": "abnormal_count",
        "mean": "abnormal_rate"
    })
)

rule_summary["abnormal_rate_percent"] = (
    rule_summary["abnormal_rate"] * 100
).round(2)

print("\n단순 Rule Score별 비정상 비율:")
display(rule_summary)

print("\nRule 기준값:")
print("High_Velocity <= ", hv_q25)
print("Cylinder_Pressure <= ", cp_q25)
print("Casting_Pressure <= ", cast_q25)
print("Cycle_Time >= ", ct_q75)
````

## 4단계: 전처리 & 피처 엔지니어링 + 상관 히트맵 + VIF

````python
# 4단계: 전처리 & 피처 엔지니어링 + 상관 히트맵 + VIF
# 목적:
# - EDA에서 확인한 속도 저하, 압력 저하, 긴 Cycle_Time 패턴을 파생변수로 만든다.
# - 모델링 전에 다중공선성을 확인한다.
# - 모델 입력용 베이스 파일을 생성한다.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.preprocessing import StandardScaler

file_path = "diecasting_model_base.csv"
df = pd.read_csv(file_path)

target_col = "Machine_Status"

# ------------------------------------------------------------
# 1. 기본 정리
# ------------------------------------------------------------
df[target_col] = df[target_col].astype(int)

id_cols = ["Shot", "_id"]
base_feature_cols = [
    col for col in df.columns
    if col not in id_cols + [target_col]
]

# 기존 결측 플래그 컬럼이 있으면 bool/int로 정리
for col in ["process_missing_flag", "all_process_missing_flag"]:
    if col in df.columns:
        df[col] = df[col].astype(int)

# ------------------------------------------------------------
# 2. Cycle_Time 극단값 처리용 파생변수
# ------------------------------------------------------------
# 원본 Cycle_Time은 남기되, 모델 안정성을 위해 log 변환 컬럼을 추가한다.
df["Cycle_Time_log1p"] = np.log1p(df["Cycle_Time"])

# 상위 1% 긴 사이클 여부
cycle_q99 = df["Cycle_Time"].quantile(0.99)
df["Cycle_Time_top1_flag"] = (df["Cycle_Time"] >= cycle_q99).astype(int)

# EDA에서 확인한 Q3 기준 긴 사이클 여부
cycle_q75 = df["Cycle_Time"].quantile(0.75)
df["Cycle_Time_long_flag"] = (df["Cycle_Time"] >= cycle_q75).astype(int)

# ------------------------------------------------------------
# 3. 속도 관련 파생변수
# ------------------------------------------------------------
df["Velocity_2_minus_1"] = df["Velocity_2"] - df["Velocity_1"]
df["Velocity_3_minus_2"] = df["Velocity_3"] - df["Velocity_2"]
df["Velocity_3_minus_1"] = df["Velocity_3"] - df["Velocity_1"]

df["Velocity_mean_123"] = df[["Velocity_1", "Velocity_2", "Velocity_3"]].mean(axis=1)
df["Velocity_std_123"] = df[["Velocity_1", "Velocity_2", "Velocity_3"]].std(axis=1)

# High_Velocity가 0인 경우는 실제 0인지 로그 오류인지 확인 필요
df["High_Velocity_zero_flag"] = (df["High_Velocity"] == 0).astype(int)

# ------------------------------------------------------------
# 4. 압력 관련 파생변수
# ------------------------------------------------------------
# 0 나눗셈 방지
eps = 1e-6

df["Casting_to_Cylinder_Pressure_Ratio"] = (
    df["Casting_Pressure"] / (df["Cylinder_Pressure"] + eps)
)

df["Pressure_Gap"] = df["Casting_Pressure"] - df["Cylinder_Pressure"]

df["Pressure_Rise_per_Cycle"] = (
    df["Pressure_Rise_Time"] / (df["Cycle_Time"] + eps)
)

# ------------------------------------------------------------
# 5. EDA 기반 Rule Flag
# ------------------------------------------------------------
hv_q25 = df["High_Velocity"].quantile(0.25)
cp_q25 = df["Cylinder_Pressure"].quantile(0.25)
cast_q25 = df["Casting_Pressure"].quantile(0.25)
ct_q75 = df["Cycle_Time"].quantile(0.75)

df["rule_low_high_velocity"] = (df["High_Velocity"] <= hv_q25).astype(int)
df["rule_low_cylinder_pressure"] = (df["Cylinder_Pressure"] <= cp_q25).astype(int)
df["rule_low_casting_pressure"] = (df["Casting_Pressure"] <= cast_q25).astype(int)
df["rule_long_cycle_time"] = (df["Cycle_Time"] >= ct_q75).astype(int)

df["rule_score"] = (
    df["rule_low_high_velocity"]
    + df["rule_low_cylinder_pressure"]
    + df["rule_low_casting_pressure"]
    + df["rule_long_cycle_time"]
)

df["rule_high_risk_flag"] = (df["rule_score"] >= 3).astype(int)

# ------------------------------------------------------------
# 6. 파생변수별 비정상 비율 점검
# ------------------------------------------------------------
print("전처리 후 데이터 크기:", df.shape)

print("\nMachine_Status 분포:")
display(
    pd.DataFrame({
        "count": df[target_col].value_counts(),
        "rate_percent": (df[target_col].value_counts(normalize=True) * 100).round(2)
    })
)

print("\nRule Score별 비정상 비율:")
rule_summary = (
    df
    .groupby("rule_score")[target_col]
    .agg(["count", "sum", "mean"])
    .rename(columns={
        "count": "row_count",
        "sum": "abnormal_count",
        "mean": "abnormal_rate"
    })
)

rule_summary["abnormal_rate_percent"] = (
    rule_summary["abnormal_rate"] * 100
).round(2)

display(rule_summary)

print("\nHigh Risk Flag별 비정상 비율:")
high_risk_summary = (
    df
    .groupby("rule_high_risk_flag")[target_col]
    .agg(["count", "sum", "mean"])
    .rename(columns={
        "count": "row_count",
        "sum": "abnormal_count",
        "mean": "abnormal_rate"
    })
)

high_risk_summary["abnormal_rate_percent"] = (
    high_risk_summary["abnormal_rate"] * 100
).round(2)

display(high_risk_summary)

# ------------------------------------------------------------
# 7. 상관 히트맵
# ------------------------------------------------------------
# 식별자 제외, 수치형 컬럼만 사용
numeric_cols = [
    col for col in df.select_dtypes(include=[np.number]).columns
    if col not in id_cols
]

corr = df[numeric_cols].corr()

plt.figure(figsize=(14, 12))
plt.imshow(corr, aspect="auto")
plt.colorbar(label="Correlation")
plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.index)), corr.index)
plt.title("Correlation Heatmap After Feature Engineering")
plt.tight_layout()
plt.show()

# 타깃과의 상관만 별도 확인
target_corr = (
    corr[target_col]
    .drop(target_col)
    .sort_values(key=lambda x: x.abs(), ascending=False)
)

print("\nMachine_Status와의 상관계수 상위 변수:")
display(target_corr.head(20).to_frame("corr_with_Machine_Status"))

# ------------------------------------------------------------
# 8. VIF 진단: statsmodels 없이 numpy로 직접 계산
# ------------------------------------------------------------
# VIF = 1 / (1 - R^2)
# 각 변수를 종속변수로 두고, 나머지 변수들로 선형회귀했을 때의 R^2를 이용한다.

def calculate_vif_numpy(X_df):
    """
    statsmodels 없이 VIF를 계산하는 함수.
    X_df: 수치형 피처만 포함된 DataFrame
    """
    X_df = X_df.copy()
    
    # 무한대 값을 결측으로 처리
    X_df = X_df.replace([np.inf, -np.inf], np.nan)
    
    # 결측이 있는 행 제거
    X_df = X_df.dropna()
    
    # 상수 컬럼 제거
    constant_cols = [
        col for col in X_df.columns
        if X_df[col].nunique() <= 1
    ]
    X_df = X_df.drop(columns=constant_cols)
    
    # 표준편차가 0인 컬럼 제거
    zero_std_cols = [
        col for col in X_df.columns
        if X_df[col].std() == 0
    ]
    X_df = X_df.drop(columns=zero_std_cols)
    
    removed_cols = constant_cols + zero_std_cols
    
    # 표준화
    X = X_df.values.astype(float)
    X_mean = X.mean(axis=0)
    X_std = X.std(axis=0)
    X_scaled = (X - X_mean) / X_std
    
    vif_rows = []
    
    for i, col in enumerate(X_df.columns):
        y = X_scaled[:, i]
        X_others = np.delete(X_scaled, i, axis=1)
        
        # 절편항 추가
        X_others = np.column_stack([
            np.ones(X_others.shape[0]),
            X_others
        ])
        
        try:
            beta, _, _, _ = np.linalg.lstsq(X_others, y, rcond=None)
            y_pred = X_others @ beta
            
            ss_res = np.sum((y - y_pred) ** 2)
            ss_tot = np.sum((y - y.mean()) ** 2)
            
            if ss_tot == 0:
                r2 = np.nan
            else:
                r2 = 1 - (ss_res / ss_tot)
            
            # R^2가 1에 매우 가까우면 VIF는 무한대에 가까움
            if pd.isna(r2):
                vif_value = np.nan
            elif r2 >= 0.999999:
                vif_value = np.inf
            else:
                vif_value = 1 / (1 - r2)
        
        except Exception:
            vif_value = np.nan
        
        vif_rows.append({
            "variable": col,
            "VIF": vif_value
        })
    
    vif_df = pd.DataFrame(vif_rows).sort_values("VIF", ascending=False)
    
    return vif_df, removed_cols


model_candidate_cols = [
    col for col in numeric_cols
    if col != target_col
]

X_vif = df[model_candidate_cols]

vif_df, removed_cols = calculate_vif_numpy(X_vif)

print("\nVIF 계산에서 제거된 상수 또는 분산 0 컬럼:")
print(removed_cols)

print("\nVIF 진단 결과:")
display(vif_df)

# ------------------------------------------------------------
# 9. VIF 10 초과 변수 확인
# ------------------------------------------------------------
high_vif = vif_df[vif_df["VIF"] > 10]

print("\nVIF 10 초과 변수:")
display(high_vif)

# ------------------------------------------------------------
# 10. 모델링용 파일 저장
# ------------------------------------------------------------
df.to_csv("diecasting_feature_engineered.csv", index=False, encoding="utf-8-sig")

print("\n저장 완료:")
print("- diecasting_feature_engineered.csv")
````

## 5단계: 모델링 3종 비교 + 5-fold CV + 임계값 튜닝 + 혼동행렬

````python
# 5단계: 모델링 3종 비교 + 5-fold CV + 임계값 튜닝 + 혼동행렬
# 목적:
# - 불균형 이진분류 문제에서 Accuracy가 아니라 PR-AUC, Recall, F1 중심으로 평가한다.
# - Logistic Regression, Random Forest, Gradient Boosting을 비교한다.
# - threshold 0.5뿐 아니라 여러 threshold에서 운영 가능한 기준을 찾는다.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import StratifiedKFold, train_test_split, cross_val_predict
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import (
    precision_score,
    recall_score,
    f1_score,
    average_precision_score,
    confusion_matrix,
    ConfusionMatrixDisplay,
    classification_report
)

file_path = "diecasting_feature_engineered.csv"
df = pd.read_csv(file_path)

target_col = "Machine_Status"

# ------------------------------------------------------------
# 1. 사용할 피처 세트 선택
# ------------------------------------------------------------
lean_features = [
    "High_Velocity",
    "Cylinder_Pressure",
    "Casting_Pressure",
    "Cycle_Time_log1p",
    "Rapid_Rise_Time",
    "Biscuit_Thickness",
    "Spray_Time"
]

operation_features = [
    "High_Velocity",
    "Cylinder_Pressure",
    "Casting_Pressure",
    "Cycle_Time_log1p",
    "Rapid_Rise_Time",
    "Biscuit_Thickness",
    "Spray_Time",
    "High_Velocity_zero_flag",
    "Cycle_Time_top1_flag",
    "rule_high_risk_flag"
]

# 우선 운영 경보형 피처 세트로 진행
feature_cols = operation_features

X = df[feature_cols].copy()
y = df[target_col].astype(int)

print("모델링 데이터 크기:", X.shape)
print("\n타깃 분포:")
display(
    pd.DataFrame({
        "count": y.value_counts(),
        "rate_percent": (y.value_counts(normalize=True) * 100).round(2)
    })
)

# ------------------------------------------------------------
# 2. 모델 정의
# ------------------------------------------------------------
models = {
    "Logistic Regression": Pipeline([
        ("imputer", SimpleImputer(strategy="median")),
        ("scaler", StandardScaler()),
        ("model", LogisticRegression(
            class_weight="balanced",
            max_iter=1000,
            random_state=42
        ))
    ]),
    
    "Random Forest": Pipeline([
        ("imputer", SimpleImputer(strategy="median")),
        ("model", RandomForestClassifier(
            n_estimators=200,
            max_depth=4,
            min_samples_leaf=10,
            class_weight="balanced",
            random_state=42,
            n_jobs=-1
        ))
    ]),
    
    "Gradient Boosting": Pipeline([
        ("imputer", SimpleImputer(strategy="median")),
        ("model", GradientBoostingClassifier(
            n_estimators=100,
            learning_rate=0.05,
            max_depth=2,
            random_state=42
        ))
    ])
}

# ------------------------------------------------------------
# 3. 5-fold CV 기반 예측 확률 생성
# ------------------------------------------------------------
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

cv_results = []

thresholds = [0.30, 0.35, 0.40, 0.45, 0.50]

oof_pred_proba_dict = {}

for model_name, model in models.items():
    print(f"\n모델 평가 중: {model_name}")
    
    # 각 행에 대해 자신이 속하지 않은 fold에서 예측한 확률
    oof_proba = cross_val_predict(
        model,
        X,
        y,
        cv=cv,
        method="predict_proba",
        n_jobs=-1
    )[:, 1]
    
    oof_pred_proba_dict[model_name] = oof_proba
    
    pr_auc = average_precision_score(y, oof_proba)
    
    for th in thresholds:
        y_pred = (oof_proba >= th).astype(int)
        
        precision = precision_score(y, y_pred, zero_division=0)
        recall = recall_score(y, y_pred, zero_division=0)
        f1 = f1_score(y, y_pred, zero_division=0)
        
        tn, fp, fn, tp = confusion_matrix(y, y_pred).ravel()
        
        cv_results.append({
            "model": model_name,
            "threshold": th,
            "PR_AUC": pr_auc,
            "precision": precision,
            "recall": recall,
            "f1": f1,
            "TN": tn,
            "FP": fp,
            "FN": fn,
            "TP": tp
        })

cv_result_df = pd.DataFrame(cv_results)

print("\n5-fold CV 기반 모델 비교 결과:")
display(
    cv_result_df.sort_values(
        ["f1", "PR_AUC"],
        ascending=False
    )
)

# ------------------------------------------------------------
# 4. 운영 관점 후보 threshold 요약
# ------------------------------------------------------------
# 조건 예시:
# - Recall 0.70 이상을 우선 만족
# - 그중 Precision과 F1이 높은 후보를 확인

candidate_df = cv_result_df[cv_result_df["recall"] >= 0.70].copy()

print("\nRecall 0.70 이상 후보:")
display(
    candidate_df.sort_values(
        ["f1", "precision"],
        ascending=False
    )
)

# ------------------------------------------------------------
# 5. 최종 후보 자동 선택
# ------------------------------------------------------------
# 우선 F1이 가장 높은 조합을 선택
best_row = cv_result_df.sort_values(
    ["f1", "PR_AUC"],
    ascending=False
).iloc[0]

best_model_name = best_row["model"]
best_threshold = best_row["threshold"]

print("\n최종 후보:")
print("모델:", best_model_name)
print("Threshold:", best_threshold)
print("PR-AUC:", round(best_row["PR_AUC"], 4))
print("Precision:", round(best_row["precision"], 4))
print("Recall:", round(best_row["recall"], 4))
print("F1:", round(best_row["f1"], 4))

# ------------------------------------------------------------
# 6. 최종 후보 혼동행렬 시각화
# ------------------------------------------------------------
best_proba = oof_pred_proba_dict[best_model_name]
best_pred = (best_proba >= best_threshold).astype(int)

cm = confusion_matrix(y, best_pred)

print("\n최종 후보 Confusion Matrix:")
print(cm)

disp = ConfusionMatrixDisplay(
    confusion_matrix=cm,
    display_labels=["Normal(0)", "Abnormal(1)"]
)

disp.plot(values_format="d")
plt.title(f"Confusion Matrix: {best_model_name} @ threshold={best_threshold}")
plt.tight_layout()
plt.show()

print("\n분류 리포트:")
print(
    classification_report(
        y,
        best_pred,
        target_names=["Normal(0)", "Abnormal(1)"],
        zero_division=0
    )
)

# ------------------------------------------------------------
# 7. 모델 비교 결과 저장
# ------------------------------------------------------------
cv_result_df.to_csv("diecasting_model_cv_results.csv", index=False, encoding="utf-8-sig")

print("\n저장 완료:")
print("- diecasting_model_cv_results.csv")
````

## 5-2단계: 룰 포함/제외 모델 성능 비교

````python
# 5-2단계: 룰 포함/제외 모델 성능 비교
# 목적:
# - rule_high_risk_flag 같은 강한 파생변수가 성능을 얼마나 끌어올렸는지 확인한다.
# - 설명형 최소 모델과 운영형 모델을 비교한다.
# - 최종 보고서에서 "룰 기반 경보 + ML 보완" 구조로 설명할 근거를 만든다.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import StratifiedKFold, cross_val_predict
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import (
    precision_score,
    recall_score,
    f1_score,
    average_precision_score,
    confusion_matrix,
    ConfusionMatrixDisplay
)

file_path = "diecasting_feature_engineered.csv"
df = pd.read_csv(file_path)

target_col = "Machine_Status"
y = df[target_col].astype(int)

lean_features = [
    "High_Velocity",
    "Cylinder_Pressure",
    "Casting_Pressure",
    "Cycle_Time_log1p",
    "Rapid_Rise_Time",
    "Biscuit_Thickness",
    "Spray_Time"
]

operation_features = [
    "High_Velocity",
    "Cylinder_Pressure",
    "Casting_Pressure",
    "Cycle_Time_log1p",
    "Rapid_Rise_Time",
    "Biscuit_Thickness",
    "Spray_Time",
    "High_Velocity_zero_flag",
    "Cycle_Time_top1_flag",
    "rule_high_risk_flag"
]

feature_sets = {
    "Lean Features": lean_features,
    "Operation Features": operation_features
}

models = {
    "Logistic Regression": Pipeline([
        ("imputer", SimpleImputer(strategy="median")),
        ("scaler", StandardScaler()),
        ("model", LogisticRegression(
            class_weight="balanced",
            max_iter=1000,
            random_state=42
        ))
    ]),
    
    "Random Forest": Pipeline([
        ("imputer", SimpleImputer(strategy="median")),
        ("model", RandomForestClassifier(
            n_estimators=200,
            max_depth=4,
            min_samples_leaf=10,
            class_weight="balanced",
            random_state=42,
            n_jobs=-1
        ))
    ]),
    
    "Gradient Boosting": Pipeline([
        ("imputer", SimpleImputer(strategy="median")),
        ("model", GradientBoostingClassifier(
            n_estimators=100,
            learning_rate=0.05,
            max_depth=2,
            random_state=42
        ))
    ])
}

thresholds = [0.30, 0.35, 0.40, 0.45, 0.50]
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

all_results = []
oof_store = {}

for feature_set_name, feature_cols in feature_sets.items():
    X = df[feature_cols].copy()
    
    for model_name, model in models.items():
        print(f"평가 중: {feature_set_name} / {model_name}")
        
        oof_proba = cross_val_predict(
            model,
            X,
            y,
            cv=cv,
            method="predict_proba",
            n_jobs=-1
        )[:, 1]
        
        key = f"{feature_set_name} / {model_name}"
        oof_store[key] = oof_proba
        
        pr_auc = average_precision_score(y, oof_proba)
        
        for th in thresholds:
            y_pred = (oof_proba >= th).astype(int)
            
            precision = precision_score(y, y_pred, zero_division=0)
            recall = recall_score(y, y_pred, zero_division=0)
            f1 = f1_score(y, y_pred, zero_division=0)
            
            tn, fp, fn, tp = confusion_matrix(y, y_pred).ravel()
            
            all_results.append({
                "feature_set": feature_set_name,
                "model": model_name,
                "threshold": th,
                "PR_AUC": pr_auc,
                "precision": precision,
                "recall": recall,
                "f1": f1,
                "TN": tn,
                "FP": fp,
                "FN": fn,
                "TP": tp
            })

result_df = pd.DataFrame(all_results)

print("\n전체 비교 결과:")
display(
    result_df.sort_values(
        ["f1", "PR_AUC"],
        ascending=False
    )
)

print("\nFeature Set별 최고 F1 결과:")
best_by_feature_set = (
    result_df
    .sort_values(["f1", "PR_AUC"], ascending=False)
    .groupby("feature_set")
    .head(1)
)

display(best_by_feature_set)

print("\nRecall 0.70 이상 후보:")
display(
    result_df[result_df["recall"] >= 0.70]
    .sort_values(["f1", "precision"], ascending=False)
)

# 최종 비교 결과 저장
result_df.to_csv(
    "diecasting_feature_set_model_comparison.csv",
    index=False,
    encoding="utf-8-sig"
)

print("\n저장 완료:")
print("- diecasting_feature_set_model_comparison.csv")
````
