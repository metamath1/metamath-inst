###  **[실행 12]**: 한 변수에 대한 품종 분포 시각화

- 이 데이터가 분류 문제로 적합한지 다시 말해 3종의 펭귄을 주어진 데이터로 잘 구별 할 수 있는지 시각화를 통해 대충 알아보고자 합니다.
- 이 목적을 위해 여러 수치형 변수의 분포를 `species`로 구분해서 시각화하고자 합니다.
- 아래 코드셀을 실행하여 시각화를 진행하세요.


```python
# 숫자형 변수를 선택하고 각 숫자형 변수에 대해서 `for` 문으로 순회하면서
# `sns.displot()`를 사용해서 분포 차트를 그려보겠습니다

# 1. `df_merged`에서 숫자형 변수를 모두 선택해 num_var에 대입합니다.
num_var = df_merged.select_dtypes(include='number')

# 2. `num_var`에 대해 `for`문을 적용하여 각 숫자형 변수에 대해서
# `sns.histlot()`으로 히스토그램을 그립니다.
for var in num_var:
    sns.histplot(data=df_merged, x=var, hue='species', kde=True)
    plt.show()
```
